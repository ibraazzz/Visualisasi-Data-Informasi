Matkul : Visualisasi Data dan Informasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Membuat data dummy untuk visualisasi
data = {
    'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'] * 3,
    'Kategori': ['Pakaian', 'Makanan', 'Elektronik'] * 12,
    'Penjualan': [100, 150, 80, 120, 160, 90, 130, 140, 85, 110, 170, 95] * 3,
    'Pendapatan': [1000, 2300, 1500, 1800, 2500, 1400, 1900, 2100, 1600, 1700, 2600, 1500] * 3
}

df = pd.DataFrame(data)

# Mengatur ukuran grid dan style
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
```

```python
# Grafik garis tren penjualan bulanan per kategori
sns.lineplot(x='Bulan', y='Penjualan', hue='Kategori', data=df, marker='o')

# Menambahkan judul dan label sumbu
plt.title('Tren Penjualan Bulanan Berdasarkan Kategori Produk', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Penjualan (Unit)', fontsize=12)
plt.legend(title='Kategori Produk')

# Tampilkan plot
plt.show()
```

```python
# Diagram batang untuk membandingkan pendapatan antar kategori
plt.figure(figsize=(10, 6))
sns.barplot(x='Kategori', y='Pendapatan', data=df, estimator=sum, ci=None)

# Menambahkan judul dan label sumbu
plt.title('Total Pendapatan Berdasarkan Kategori Produk', fontsize=16)
plt.xlabel('Kategori Produk', fontsize=12)
plt.ylabel('Total Pendapatan (Mata Uang)', fontsize=12)

# Tampilkan plot
plt.show()
```
