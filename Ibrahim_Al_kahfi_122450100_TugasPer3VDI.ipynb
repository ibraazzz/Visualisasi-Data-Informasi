# 1. Menentukan Data untuk Visualisasi
Saya akan menggunakan dataset sederhana yang berisi penjualan produk per bulan untuk visualisasi. Dataset ini bisa terdiri dari beberapa variabel berikut:

Bulan: Nama bulan dalam setahun

Penjualan Produk A: Jumlah penjualan produk A

Penjualan Produk B: Jumlah penjualan produk B

# 2. Definisi Tujuan, User, Fungsi, dan Tone Visualisasi
Tujuan: Tujuan visualisasi ini adalah untuk menganalisis tren penjualan produk A dan B sepanjang tahun serta membandingkan performa keduanya.

User: Pengguna yang dituju adalah tim manajemen penjualan atau tim marketing yang ingin memahami fluktuasi penjualan dan memutuskan strategi bisnis berdasarkan data.

Fungsi: Fungsi utama dari visualisasi ini adalah monitoring dan analisis kinerja penjualan, membantu mengambil keputusan strategis berdasarkan tren.

Tone: Visualisasi ini akan dibuat dengan tone yang profesional, bersih, dan informasi langsung, karena target audiensnya adalah para pengambil keputusan di perusahaan.

# 3. Membuat Visualisasi
Saya akan membuat dua jenis visualisasi:

Line Plot untuk menunjukkan tren penjualan sepanjang tahun.

Bar Plot untuk membandingkan jumlah penjualan produk A dan B di setiap bulan.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data
data = {
    'Bulan': ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli',
              'Agustus', 'September', 'Oktober', 'November', 'Desember'],
    'Penjualan Produk A': [500, 450, 600, 700, 750, 800, 820, 900, 950, 1000, 1050, 1100],
    'Penjualan Produk B': [300, 350, 400, 500, 550, 600, 620, 700, 750, 800, 850, 900]
}

# Membuat DataFrame
df = pd.DataFrame(data)

# Set style visualisasi
sns.set(style="whitegrid")

# Membuat line plot untuk melihat tren penjualan
plt.figure(figsize=(10, 6))
plt.plot(df['Bulan'], df['Penjualan Produk A'], marker='o', label='Produk A', color='b')
plt.plot(df['Bulan'], df['Penjualan Produk B'], marker='s', label='Produk B', color='r')
plt.title('Tren Penjualan Produk A dan B Sepanjang Tahun', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Jumlah Penjualan', fontsize=12)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()

# Menampilkan line plot
plt.show()

# Membuat bar plot untuk perbandingan penjualan produk A dan B
plt.figure(figsize=(10, 6))
df_melted = df.melt(id_vars="Bulan", value_vars=['Penjualan Produk A', 'Penjualan Produk B'],
                    var_name="Produk", value_name="Penjualan")
sns.barplot(x='Bulan', y='Penjualan', hue='Produk', data=df_melted)
plt.title('Perbandingan Penjualan Produk A dan B', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Jumlah Penjualan', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Menampilkan bar plot
plt.show()
```
